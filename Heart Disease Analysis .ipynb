```python
#Checking the chest pain type 
#4 values 0 = typical angina ,1 = atypical angina ,2 = non-anginal pain,3 = asymptomatic 
```

```python
sns.countplot(x='cp', data=data)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.show
```

```python
#as we can see typical angina is more common between people in this study
```

```python
pain_distribute = data.groupby(['cp'], as_index = False)['target'].sum().sort_values(by='target', ascending = False)
sns.barplot(x='cp', y='target', data = pain_distribute)
```

```python
sns.countplot(x='cp', hue='target', data = data)
plt.legend(labels = ['No disease','Disease'])
```

```python
sns.countplot(x='fbs', hue='target', data = data)
plt.legend(labels = ['No disease','Disease'])
```
